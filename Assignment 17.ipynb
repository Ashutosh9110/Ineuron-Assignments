```python
# 1 Assign the value 7 to the variable guess_me. Then, write the conditional tests (if,
# else, and elif) to print the string 'too low' if guess_me is less than 7, 'too high' if
# greater than 7, and 'just right' if equal to 7.


guess_me = 7
if guess_me < 7:
    print("too low")
elif guess_me > 7:
    print("too high")
else:
    print("just right")
```

```python
# 4. Use a list comprehension to make a list of the even numbers in range(10)

# Ans: 

even_numbers = [num for num in range(10) if num % 2 == 0]
print(even_numbers)
```

```python
# 5. Use a dictionary comprehension to create the dictionary squares. Use range(10) to return the
# keys, and use the square of each key as its value.

# Ans: 
    
limit = 10
squares = {num: num * num for num in range(limit)}
print(squares)
```

```python
# 6. Construct the set odd from the odd numbers in the range using a set comprehension (10).

# Ans: 
    
limit = 10
odd = {num for num in range(limit) if num % 2 == 1}
print(odd)
```

```python
# 7. Use a generator comprehension to return the string &#39;Got &#39; and a number for the numbers in
# range(10). Iterate through this by using a for loop.


# Ans: 
limit = 10
string_generator = ('Got ' + str(num) for num in range(limit))
for item in string_generator:
    print(item)
```

```python
# 9. Define a generator function called get_odds that returns the odd numbers from range(10). Use a
# for loop to find and print the third value returned.

# Ans: 

limit = 10
get_odds = (num for num in range(limit) if not num % 2 == 0)
count = 0
for num in get_odds:
    if count == 2:
        print(num)
        break
    count += 1
```

```python
# 10: Define an exception called OopsException. Raise this exception to see what hap‚Äê
# pens. Then write the code to catch this exception and print 'Caught an oops'.

# Ans: 
    
class OopsException(Exception):
    pass


def with_exception(a):
    if a < 0:
        raise OopsException(a)


try:
    with_exception(-1)
except OopsException as err:
    print('Caught an oops')
```

```python
# 11: Use zip() to make a dictionary called movies that pairs these lists: titles =
# ['Creature of Habit', 'Crewel Fate'] and plots = ['A nun turns into a mon
# ster', 'A haunted yarn shop'].

# Ans: 

titles = ['Creature of Habit', 'Crewel Fate']
plots = ['A nun turns into a monster', 'A haunted yarn shop']

movies = {}
for title, plot in zip(titles, plots):
    movies[title] = plot
# or movies = dict(zip(titles, plots))
print(movies)
```
